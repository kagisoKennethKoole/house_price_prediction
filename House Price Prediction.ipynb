```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# creating artificial data to forecast home prices
np.random.seed(42)
predictedPrice = np.random.randn(1000, 5)
actualPrice = np.random.randn(1000) * 20  # Adding some noise

# dividing the data into sets for testing and training
predictedPrice_train, predictedPrice_test, actualPrice_train, actualPrice_test = train_test_split(predictedPrice, actualPrice, test_size=0.2, random_state=42)

# Setting up the model for linear regression
model = LinearRegression()

# Using the training data to train the model
model.fit(predictedPrice_train, actualPrice_train)

# predicting using the test set
predictions = model.predict(predictedPrice_test)

# Analyzing the model
mse = mean_squared_error(actualPrice_test, predictions)

# Plotting actual vs the predicted prices
plt.figure(figsize=(8, 6))
plt.scatter(actualPrice_test, predictions, color='blue', alpha=0.5, label='Predicted')
plt.scatter(actualPrice_test, actualPrice_test, color='red', alpha=0.5, label='Actual')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.legend()
plt.title("Actual vs the Predicted House Prices")
plt.show()



```
